```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
short = 0.03
long = 0.05
b = long/short

p = 0.4
a = 25216
f = (b*p-(1-p))/b
bid_price = 71.8


print(f'f: {f}, total price: {a*f}, share: {a*f/bid_price}')
print(f'close price:{bid_price*(1+long)}, close price:{bid_price*(1-short)}')
```

```python
def kali(buy, long):
    win_rate = np.arange(1,11)/10   
    short = np.arange(1,101, 4)/1000
    
    fig, axs = plt.subplots()
    for p in win_rate:
        output = []
        for short_ in short:
            b = long/short_
            f = (b*p-(1-p))/b
            output.append(f)
    
        
        axs.plot(short*100, np.array(output), 'o', label=str(p))   
        axs.set_xlim(0.5, 10)
        axs.set_ylim(-1, 1)   
    
    axs.set_xlabel('short(%)')
    axs.set_ylabel('f')
    axs.grid(True)
    plt.legend(title='Parameter where:')
    fig.tight_layout()
    plt.show()            
    
    return 
```

```python
kali(bid_price, 0.1)
```
